```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
matches=pd.read_csv("matches.csv")
delivery=pd.read_csv("deliveries.csv")
```

```python
 total_score=delivery.groupby(["match_id","inning"]).sum()["total_runs"].reset_index()
```

```python
total_score=total_score[total_score["inning"]==1]
```

```python
total_score.head()
```

```python
match_df=matches.merge(total_score[["match_id","total_runs"]],left_on="id",right_on="match_id")
```

```python
match_df["team1"].unique()
```

```python
teams=['Sunrisers Hyderabad', 'Mumbai Indians', 'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals',
       'Delhi Capitals']
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df=match_df[match_df["team1"].isin(teams)]
match_df=match_df[match_df["team2"].isin(teams)]
```

```python
match_df.shape
```

```python
match_df.head()
```

```python
match_df[match_df["dl_applied"]==0]
```

```python
match_df=match_df[["match_id","city","winner","total_runs"]]
```

```python
match_df.head()
```

```python
delivery=match_df.merge(delivery,on="match_id")
```

```python
delivery.head()
```

```python
delivery=delivery[delivery["inning"]==2]
```

```python
delivery.shape
```

```python
delivery["current_score"]=delivery.groupby("match_id").cumsum()["total_runs_y"]
```

```python
delivery['runs_left'] = delivery['total_runs_x'] - delivery['current_score']
```

```python
delivery["balls_left"]=120-((delivery["over"]-1)*6+delivery["ball"])
```

```python
# crr = runs/overs
delivery['crr'] = (delivery['current_score']*6)/(120 - delivery['balls_left'])
```

```python
delivery['rrr'] = (delivery['runs_left']*6)/delivery['balls_left']
```

```python
def result(x):
    return 1 if x['batting_team'] == x['winner'] else 0
```

```python
delivery['result'] = delivery.apply(result,axis=1)
```

```python
delivery.head(10)
```

```python
delivery["player_dismissed"]=delivery["player_dismissed"].fillna("0")
```

```python
delivery["player_dismissed"]=delivery["player_dismissed"].apply(lambda x:x if x=="0"else"1")
delivery["player_dismissed"]=delivery["player_dismissed"].astype("int64")
```

```python
delivery["player_dismissed"]=delivery.groupby("match_id").cumsum()["player_dismissed"]
```

```python
delivery["wicket_left"]=10-delivery["player_dismissed"]
```

```python
delivery.head()
```

```python
final_df = delivery[['batting_team','bowling_team','city','runs_left','balls_left','wicket_left','total_runs_x','crr','rrr','result']]
```

```python
final_df=final_df[final_df["balls_left"]!=0]
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
x=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trn=ColumnTransformer([
    ("trf",OneHotEncoder(sparse_output=False,drop="first"),["batting_team","bowling_team","city"])
],remainder="passthrough"
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[
    ("step1",trn),
    ("step2",LogisticRegression(solver="liblinear"))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
